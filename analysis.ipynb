```python
import pandas as pd
import matplotlib.pyplot as plt
"""charger un dataset médical depuis sklearn"""
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
"""convertir en Dataframe"""
df = pd.DataFrame(data.data, columns=data.feature_names)
"""ajouter la colonne diagnostic"""
df["diagnostic"] = data.target
"""afficher les 5 premières lignes"""
df.head()
```

```python
"""histogramme d'une variable biologigue"""
plt.hist(df["mean radius"], bins=20)
plt.xlabel("Rayon moyen")
plt.ylabel("Nombre de patients")
plt.title("Distribution du rayon moyen des cellules")
plt.show()
```
