```python
import pandas as pd
import matplotlib.pyplot as plt
"""charger un dataset médical depuis sklearn"""
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
"""convertir en Dataframe"""
df = pd.DataFrame(data.data, columns=data.feature_names)
"""ajouter la colonne diagnostic"""
df["diagnostic"] = data.target
"""afficher les 5 premières lignes"""
df.head()
```

```python
""" 0=cancer malin 1=cancer bénin"""
"""dimension du dataset"""
df.shape
```

```python
"""information générale"""
df.info()
```

```python
"""statistiques descriptives"""
df.describe()
```

```python
"""compter les cas"""
df["diagnostic"].value_counts()
```

```python
"""histogramme d'une variable biologigue"""
plt.hist(df["mean radius"], bins=20)
plt.xlabel("Rayon moyen")
plt.ylabel("Nombre de patients")
plt.title("Distribution du rayon moyen des cellules")
plt.show()
```

```python
"""séparer les cellules x et y"""
x = df.drop("diagnostic" , axis=1)
y = df["diagnostic"]

print(x.shape)
print(y.shape)
```

```python
"""cellule train/test"""
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(y_test.shape)
```

```python
"""model ML"""
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
"""prédictions"""
from sklearn.metrics import accuracy_score, confusion_matrix
y_pred = model.predict(x_test)
print("Précision :", accuracy_score(y_test, y_pred))
```

```python
"""Matrice de confusion"""
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
"""visualiser les matrices"""
import seaborn as sns

plt.figure()
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("prédictions")
plt.ylabel("Valeurs réelles")
plt.title("Matrice de confusion")
plt.show()
```
